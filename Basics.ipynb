## 6. Write user defined function

### 6.1 Copy range, row, column, cell ...

```python
# write a function to copy a range

def copy_range(sheet_from, cellrange, sheet_to, row_no, col_no ):
    cell_range = sheet_from[cellrange]
    col_counter = col_no
    for row in cell_range:
        for cell in row:
            sheet_to.cell(row=row_no, column=col_counter).value= cell.value
            col_counter = col_counter +1
        row_no = row_no + 1
        col_counter = col_no
```

```python
# create a new sheet were to copy
wb.create_sheet(title='Department_UTD')
departsheet= wb['Department_UTD']

# copy the wanted range: row
copy_range(sheet_from = sheet,
           cellrange='D1:Q5', 
           sheet_to = departsheet,
           row_no = 1, 
           col_no = 1)

# save workbook 
wb.save('Budget_1.xlsx')
```

### Rewrite a cell range

```python
#Takes: start cell, end cell, and sheet you want to copy from.
def copyRange(startCol, startRow, endCol, endRow, sheet):
    rangeSelected = []
    #Loops through selected Rows
    for i in range(startRow,endRow + 1,1):
        #Appends the row to a RowSelected list
        rowSelected = []
        for j in range(startCol,endCol+1,1):
            rowSelected.append(sheet.cell(row = i, column = j).value)
        #Adds the RowSelected List and nests inside the rangeSelected
        rangeSelected.append(rowSelected)

    return rangeSelected
```

```python
#Paste range
#Paste data from copyRange into template sheet
def pasteRange(startCol, startRow, endCol, endRow, sheetReceiving,copiedData):
    countRow = 0
    for i in range(startRow,endRow+1,1):
        countCol = 0
        for j in range(startCol,endCol+1,1):
            
            sheetReceiving.cell(row = i, column = j).value = copiedData[countRow][countCol]
            countCol += 1
        countRow += 1
```

```python
# column totals

first_row = 2
last_row = ws.max_row
end_col = ws.max_column
sum_col = end_col + 1
start_col = 2


for row in ws.iter_rows(min_row=2, 
                        max_row=last_row, 
                        min_col=  sum_col, 
                        max_col= sum_col):
    for cell in row:
        cell_sum_start = str(get_column_letter(start_col)) + str(first_row)
        cell_col_end = str(get_column_letter(end_col)) + str(first_row)
        cell.value = '=SUM({0}:{1})'.format(cell_sum_start, cell_col_end) 
        first_row= first_row+1

wb.save('Groceries_1.xlsx')
```

```python
# row totals

first_row = 2
last_row = ws.max_row
sum_row = last_row + 1
start_col = 2
end_col = ws.max_column
for row in ws.iter_rows(min_row=sum_row,
                        max_row=sum_row, 
                        min_col=start_col,
                        max_col=end_col):
    for cell in row:
        cell_sum_start = cell.column_letter + str(first_row)
        cell_sum_end = cell.column_letter + str(last_row)
        cell.value = '=SUM({0}:{1})'.format(cell_sum_start, cell_sum_end) 

wb.save('Groceries_1.xlsx')
```

```python
import xlsxwriter
from xlsxwriter.utility import xl_rowcol_to_cell
from xlsxwriter.utility import xl_col_to_name
from xlsxwriter.utility import xl_range, xl_range_abs
from xlsxwriter.utility import xl_rowcol_to_cell
```

```python
# Load raw in the workbook
wb = load_workbook('Groceries_1.xlsx') # load processed file with 16 excels

# Get sheet names
print(wb.sheetnames) # rename 2nd sheet previously to Headcount

sheet = wb['Groceries_List']

# Print the sheet title
print('Sheet Title:',sheet.title)

# Convert Sheet to DataFrame 
df = pd.DataFrame(sheet.values)
df.head()
```

```python
headers = df.iloc[0,:]
df.columns = headers 
df = df.loc[1:]
df = df.drop([2])

print(df.shape)
df.head()
```

```python
# copute the range of the whole dataframe 
max_row, max_col = df.shape
print(max_row, max_col)

df_range = xl_range(0,0, max_row, max_col - 1)
print(df_range)
```

```python
# start indices
start = xl_rowcol_to_cell(0,0)
start_col = xl_col_to_name(0)
start_row = 0
print("start point:", start, 
      "start_row:", start_row,      
      "start col:", start_col)
```

```python
# end indices
end_point = xl_rowcol_to_cell(max_row, max_col-1)
end_col = xl_col_to_name(max_col-1)
end_row = max_row + 1
print("end point:", end_point, 
      "end_row:", end_row,      
      "end col:", end_col)
```

```python
writer = pd.ExcelWriter('Groceries_Formated.xlsx', engine='xlsxwriter',
                        datetime_format='d mmm yy', 
                        date_format='d mmm yy',)
df.to_excel(writer, sheet_name= 'Groceries' , 
            index=False,
            startrow=0,
            startcol=0)
workbook = writer.book
worksheet = writer.sheets['Groceries']
#  resize the sheet by adjusting the zoom
worksheet.set_zoom(60)
```

```python
# Add a header format.
header_format = workbook.add_format({
    'bold': True,
    'text_wrap': True,
    'valign': 'top',
    'fg_color': '#D7E4BC',
    'border': 1})

# Write the column headers with the defined format.
for col_num, value in enumerate(df.columns.values):
    worksheet.write(0, col_num, value, header_format)

```
